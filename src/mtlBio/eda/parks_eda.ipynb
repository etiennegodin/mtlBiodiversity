```python

x = gdf['Type'].unique()
gdf = gdf[(gdf['TYPO1'] == 'Grand parc')|
            (gdf['TYPO2'] == 'Parc-nature') |
            (gdf['TYPO2'] == 'Îlot de verdure') |
            (gdf['TYPO2'] == 'Parc de quartier') |
            (gdf['TYPO2'] == 'Jardin Botanique') |
            (gdf['TYPO2'] == 'Bande riveraine') |
            (gdf['Type'] == 'Cimetière') |
            (gdf['Nom'] == 'Jarry') |
            (gdf['Nom'] == 'Lieu Historique National du Canada du-Canal-de-Lachine') 
              ]
gdf.reset_index(inplace=True, drop=True)

gdf.head()
```

```python
#Merge parks with same name 
gdf = gdf.dissolve(by='Nom', as_index=False, aggfunc= 'first')
print(gdf.size)
gdf.head()
```
